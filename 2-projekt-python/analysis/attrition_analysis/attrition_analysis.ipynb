### PowyÅ¼sze wartoÅ›ci wyraÅ¼one w %

```python
percent_job_level = []

for i in range(0, len(attrition_job_levels)):
    total = sum(list(attrition_job_levels[i].values()))
    percent_job_level.append(
        {
        'JobLevel': i + 1,
        'Attrition': 0,
        'Percent': round(list(attrition_job_levels[i].values())[0] / total * 100, 2)
        })
    percent_job_level.append(
        {
        'JobLevel': i + 1,
        'Attrition': 1,
        'Percent': round(list(attrition_job_levels[i].values())[1] / total * 100, 2)
        })

percent_job_level
```
