# Zamiana wartości zmiennych ze string na int

```python
#aby ręcznie nie przepisywać danych i nie tworzyć kolumn za pomocą funkcji get dummies instaluję poniższy pakiet, który automatycznie poprzypisuje mi zmienne int do zmiennych str
le = df.copy()
from sklearn.preprocessing import LabelEncoder

#tworzę funkcję, która macierz zaweirającą nazwy kolumn i zakdowane zmienne
def encode_features(le, encoder=LabelEncoder):
    acc = []
    for col in df.columns:
        fitted = encoder().fit(le[col].values)
        le[col] = fitted.transform(le[col].values)
        dic = (col, fitted)
        acc.append(dic)
    return acc, edf

#tworzę funkcję, aby moc odszyforwać jak zostały zakodowane zmienne
def get_encoder(qcol, encoders):
    for col, encoder in encoders:
        if qcol == col:
            return encoder
    return None
```

```python
#przystępuję do zakodowania naszego data frame'u

encoders, edf = encode_features(le)
edf
```

```python
#w razie potrzeby kod do rozszyforwania na przykładzie kolumny 'cap-shape'
# wywołujemy encodery jakie były przypisane w tej kolumnie
cap_shape_encoder = get_encoder('cap-shape', encoders)
# wywołujemy wartości encoderów przypisanych w tej kolumnie
encoded_values_of_cap_shape = edf['cap-shape'].values
#rozszyfrowanie encodera
original_values_of_cap_shape =  cap_shape_encoder.inverse_transform(encoded_values_of_cap_shape)

pd.DataFrame({'Encoded': encoded_values_of_cap_shape, 'Original': original_values_of_cap_shape}).head()
```
